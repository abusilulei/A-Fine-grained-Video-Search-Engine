```python
import pandas as pd
```

```python
raw_data = pd.read_json('./results_2.json')
raw_data
```

```python
# drop column "typenames"
raw_data_normalized = raw_data.drop(columns=['content_category', 'upics'])
raw_data_normalized.head(2)

# rename columns, make "pic" -> "upics"
raw_data_normalized = raw_data_normalized.rename(columns={'pic': 'upics'})
raw_data_normalized
```

```python
# each element in the column "subtitles" is a list of dictionaries, like `[{'start_time': 0.04, 'contents': '这位尊敬的委员'},...]`

# we need to shorten the keys from "start_time" to "s", and the "contents" to "c". 

# write a function and apply it to the column "subtitles".

def convert_subtitles(subtitles):
    for i in range(len(subtitles)):
        subtitles[i] = {'s': subtitles[i]['start_time'], 'c': subtitles[i]['contents']}
    return subtitles

raw_data_normalized['subtitles'] = raw_data['subtitles'].apply(convert_subtitles)
raw_data_normalized
```

```python
# for each unique typenames, assign a unique id. 
# change the original typenames to the new id.
# and create a new table for "typenames".

typenames = raw_data['typenames'].unique()
type_id = range(1, len(typenames)+1)
typenames_dict = dict(zip(typenames, type_id))

# typenames_dict

raw_data_normalized = raw_data_normalized.copy()
raw_data_normalized['type_id'] = raw_data['typenames'].map(typenames_dict)
raw_data_normalized = raw_data_normalized.drop(columns=['typenames'])

# similarly, for "lan"
lan = raw_data['lan'].unique()
lan_id = range(1, len(lan)+1)
lan_dict = dict(zip(lan, lan_id))
raw_data_normalized['lan_id'] = raw_data['lan'].map(lan_dict)
raw_data_normalized = raw_data_normalized.drop(columns=['lan'])

raw_data_normalized.head(2)
```

```python
# create two new tables for "typenames" and "lan"
typenames_table = pd.DataFrame({'type_id': type_id, 'typenames': typenames})
lan_table = pd.DataFrame({'lan_id': lan_id, 'lan': lan})
typenames_table
```

```python
lan_table
```

```python
# keep only one row for rows with the same column "bvids" value
raw_data_normalized = raw_data_normalized.drop_duplicates(subset=['bvids'], keep='first')
raw_data_normalized
```

```python
# keep only one row for rows with the same column "bvids" value
raw_data_normalized = raw_data_normalized.drop_duplicates(subset=['subtitles'], keep='first')
raw_data_normalized
```

```python
# export the normalized data and the new tables to csv files
raw_data_normalized.to_json('./output/info.json', orient='records', force_ascii=False)
typenames_table.to_csv('./output/typenames.csv', index=False)
lan_table.to_csv('./output/lan.csv', index=False)
```
